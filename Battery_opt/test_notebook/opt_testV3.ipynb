```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory

# ==============================
# Model parameters
# ==============================
T = 24  # Number of hours (e.g., 24 for a single day)
P_max = 10.0   # Maximum power capacity (MW)
E_max = 20.0   # Maximum energy capacity (MWh)
E_min = 2.0    # Minimum energy capacity (MWh)
eta_discharge = 0.95  # Discharge efficiency
eta_charge = 0.95     # Charge efficiency
E_initial = 10.0      # Initial state of charge (MWh)
delta_t = 1.0         # Time step in hours

# Example price arrays: real-time, day-ahead, ancillary.
# (Ensure these arrays each have length T.)
price_rt = DAM_one.SP_Price_Houston.values      # Real-time prices ($/MWh)
price_da = DAM_one.DA_Price_Houston.values      # Day-ahead prices ($/MWh)
price_reg_up = ancilary_service_one.REGUP.values
price_reg_down = ancilary_service_one.REGDN.values
price_rrs = ancilary_service_one.RRS.values
price_ecrs = ancilary_service_one.ECRS.values

# ==============================
# NEW: Basic data checks (Suggestion #4)
# ==============================
assert len(price_rt) == T,      "Real-time price array must be length T."
assert len(price_da) == T,      "Day-ahead price array must be length T."
assert len(price_reg_up) == T,  "RegUp price array must be length T."
assert len(price_reg_down) == T,"RegDown price array must be length T."
assert len(price_rrs) == T,     "RRS price array must be length T."
assert len(price_ecrs) == T,    "ECRS price array must be length T."

# ==============================
# Create model
# ==============================
model = pyo.ConcreteModel()

# Sets
model.T = pyo.RangeSet(0, T-1)  # 0..23 if T=24

# ==============================
# Decision variables
# ==============================
model.P_discharge = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
model.P_charge    = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
model.P_rt        = pyo.Var(model.T, domain=pyo.Reals, bounds=(-P_max, P_max))
model.E_da        = pyo.Var(model.T, domain=pyo.Reals)

model.R_up   = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
model.R_down = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
model.R_rrs  = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
model.R_ecrs = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))

model.E_bat  = pyo.Var(model.T, domain=pyo.Reals, bounds=(E_min, E_max))

# Fix initial SOC
model.E_bat[0].fix(E_initial)

# ==============================
# Constraints
# ==============================

# (1) Real-time power definition
def p_rt_definition(m, t):
    return m.P_rt[t] == m.P_discharge[t] - m.P_charge[t]
model.p_rt_def = pyo.Constraint(model.T, rule=p_rt_definition)

# (2) Energy balance
def energy_balance(m, t):
    if t == 0:
        return pyo.Constraint.Skip
    return m.E_bat[t] == (
        m.E_bat[t-1]
        - (m.P_discharge[t] * delta_t) / eta_discharge
        + (m.P_charge[t]    * delta_t) * eta_charge
    )
model.energy_balance = pyo.Constraint(model.T, rule=energy_balance)

# (3) Maximum power constraint
def power_max_constraint(m, t):
    return m.P_rt[t] + m.R_up[t] + m.R_rrs[t] + m.R_ecrs[t] <= P_max
model.power_max = pyo.Constraint(model.T, rule=power_max_constraint)

# (4) Minimum power constraint
def power_min_constraint(m, t):
    return m.P_rt[t] - m.R_down[t] >= -P_max
model.power_min = pyo.Constraint(model.T, rule=power_min_constraint)

# (5) ECRS energy constraint
def ecrs_energy_constraint(m, t):
    return m.E_bat[t] >= (m.R_ecrs[t] * 2.0 * delta_t) / eta_discharge
model.ecrs_energy = pyo.Constraint(model.T, rule=ecrs_energy_constraint)

# (6) Bounds on day-ahead commitment
for t in model.T:
    model.E_da[t].setlb(-P_max * delta_t)
    model.E_da[t].setub(P_max  * delta_t)

# (7) Optional link day-ahead MWh to actual physical discharge
def couple_da_discharge(m, t):
    return m.E_da[t] <= m.P_discharge[t] * delta_t
model.couple_da_discharge = pyo.Constraint(model.T, rule=couple_da_discharge)

# ==============================
# NEW: RegUp / RegDown feasibility constraints (Suggestion #1)
# ==============================
def reg_up_feasibility(m, t):
    # Ensure there's headroom to go "up" if called
    # i.e. can't be discharging at Pmax already
    return m.P_discharge[t] + m.R_up[t] <= P_max
model.reg_up_feasibility = pyo.Constraint(model.T, rule=reg_up_feasibility)

def reg_down_feasibility(m, t):
    # Ensure there's headroom to go "down" if called
    # i.e. can't be charging at Pmax already
    return m.P_charge[t] + m.R_down[t] <= P_max
model.reg_down_feasibility = pyo.Constraint(model.T, rule=reg_down_feasibility)

# Also check that we have enough energy to provide RegUp ...
def reg_up_energy(m, t):
    # Battery must have enough SOC to sustain R_up for 1 hour if called
    return m.E_bat[t] >= (m.R_up[t] * delta_t) / eta_discharge
model.reg_up_energy = pyo.Constraint(model.T, rule=reg_up_energy)

# ... and enough 'free capacity' to absorb RegDown if needed
def reg_down_energy(m, t):
    # Battery must have enough headroom to store R_down if called
    return (E_max - m.E_bat[t]) >= (m.R_down[t] * delta_t) * eta_charge
model.reg_down_energy = pyo.Constraint(model.T, rule=reg_down_energy)

# ==============================
# NEW: Final SOC requirement (Suggestion #3)
# ==============================
# Example: force final SOC to match the initial SOC
model.E_bat[T-1].fix(E_initial)

# ==============================
# Objective
# ==============================
def obj_rule(m):
    return sum(
        m.P_rt[t] * price_rt[t] * delta_t
        + m.E_da[t] * (price_da[t] - price_rt[t])
        + m.R_up[t]   * price_reg_up[t]
        + m.R_down[t] * price_reg_down[t]
        + m.R_rrs[t]  * price_rrs[t]
        + m.R_ecrs[t] * price_ecrs[t]
        for t in m.T
    )
model.obj = pyo.Objective(rule=obj_rule, sense=pyo.maximize)

# ==============================
# Solve
# ==============================
solver = SolverFactory('glpk', executable='/usr/local/bin/glpsol')
results = solver.solve(model, tee=True)

status = results.solver.status
term_cond = results.solver.termination_condition

if (status != pyo.SolverStatus.ok) or (term_cond != pyo.TerminationCondition.optimal):
    print(f"\nNo feasible (or optimal) solution found.\n"
          f"Solver Status: {status}\n"
          f"Termination Condition: {term_cond}")
else:
    print("\nOptimal solution found.\n")
    for t in model.T:
        print(f"Hour {t:2d}: "
              f"P_rt    = {pyo.value(model.P_rt[t]):6.2f}, "
              f"E_da    = {pyo.value(model.E_da[t]):6.2f}, "
              f"R_up    = {pyo.value(model.R_up[t]):6.2f}, "
              f"R_down  = {pyo.value(model.R_down[t]):6.2f}, "
              f"R_rrs   = {pyo.value(model.R_rrs[t]):6.2f}, "
              f"R_ecrs  = {pyo.value(model.R_ecrs[t]):6.2f}, "
              f"E_bat   = {pyo.value(model.E_bat[t]):6.2f}")

    total_profit = pyo.value(model.obj)
    print(f"\nTotal 24-hour profit = ${total_profit:,.2f}")
```

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory

def optimize_battery_schedule(E_initial, price_rt, price_da, price_reg_up, price_reg_down, price_rrs, price_ecrs):
    """
    Optimize a battery's daily market participation and return the 24-hour profit and final SOC.

    Parameters:
    - price_rt: List of real-time prices ($/MWh) [length = 24]
    - price_da: List of day-ahead prices ($/MWh) [length = 24]
    - price_reg_up: List of Regulation Up prices ($/MW) [length = 24]
    - price_reg_down: List of Regulation Down prices ($/MW) [length = 24]
    - price_rrs: List of RRS prices ($/MW) [length = 24]
    - price_ecrs: List of ECRS prices ($/MW) [length = 24]

    Returns:
    - total_profit: The optimal 24-hour profit ($).
    - final_soc: The battery's final state of charge (MWh).
    """

    # ==============================
    # Model parameters
    # ==============================
    T = 24  # Number of hours
    P_max = 10.0   # Maximum power capacity (MW)
    E_max = 20.0   # Maximum energy capacity (MWh)
    E_min = 2.0    # Minimum energy capacity (MWh)
    eta_discharge = 0.95  # Discharge efficiency
    eta_charge = 0.95     # Charge efficiency
    delta_t = 1.0         # Time step in hours

    # ==============================
    # Data validation
    # ==============================
    assert len(price_rt) == T,      "Real-time price array must be length 24."
    assert len(price_da) == T,      "Day-ahead price array must be length 24."
    assert len(price_reg_up) == T,  "RegUp price array must be length 24."
    assert len(price_reg_down) == T,"RegDown price array must be length 24."
    assert len(price_rrs) == T,     "RRS price array must be length 24."
    assert len(price_ecrs) == T,    "ECRS price array must be length 24."

    # ==============================
    # Create optimization model
    # ==============================
    model = pyo.ConcreteModel()

    # Time set
    model.T = pyo.RangeSet(0, T-1)

    # Decision variables
    model.P_discharge = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
    model.P_charge    = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
    model.P_rt        = pyo.Var(model.T, domain=pyo.Reals, bounds=(-P_max, P_max))
    model.E_da        = pyo.Var(model.T, domain=pyo.Reals)

    model.R_up   = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
    model.R_down = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
    model.R_rrs  = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))
    model.R_ecrs = pyo.Var(model.T, domain=pyo.NonNegativeReals, bounds=(0, P_max))

    model.E_bat  = pyo.Var(model.T, domain=pyo.Reals, bounds=(E_min, E_max))

    # Fix initial SOC
    model.E_bat[0].fix(E_initial)

    # ==============================
    # Constraints
    # ==============================
    # Real-time power definition
    def p_rt_definition(m, t):
        return m.P_rt[t] == m.P_discharge[t] - m.P_charge[t]
    model.p_rt_def = pyo.Constraint(model.T, rule=p_rt_definition)

    # Energy balance
    def energy_balance(m, t):
        if t == 0:
            return pyo.Constraint.Skip
        return m.E_bat[t] == (
            m.E_bat[t-1]
            - (m.P_discharge[t] * delta_t) / eta_discharge
            + (m.P_charge[t]    * delta_t) * eta_charge
        )
    model.energy_balance = pyo.Constraint(model.T, rule=energy_balance)

    # Maximum and minimum power constraints
    model.power_max = pyo.Constraint(model.T, rule=lambda m, t: m.P_rt[t] + m.R_up[t] + m.R_rrs[t] + m.R_ecrs[t] <= P_max)
    model.power_min = pyo.Constraint(model.T, rule=lambda m, t: m.P_rt[t] - m.R_down[t] >= -P_max)

    # Ensure sufficient energy for ECRS
    model.ecrs_energy = pyo.Constraint(model.T, rule=lambda m, t: m.E_bat[t] >= (m.R_ecrs[t] * 2.0 * delta_t) / eta_discharge)

    # Bounds on day-ahead commitment
    for t in model.T:
        model.E_da[t].setlb(-P_max * delta_t)
        model.E_da[t].setub(P_max  * delta_t)

    # Optional: Link day-ahead MWh to actual discharge
    model.couple_da_discharge = pyo.Constraint(model.T, rule=lambda m, t: m.E_da[t] <= m.P_discharge[t] * delta_t)

    # RegUp/RegDown feasibility constraints
    model.reg_up_feasibility = pyo.Constraint(model.T, rule=lambda m, t: m.P_discharge[t] + m.R_up[t] <= P_max)
    model.reg_down_feasibility = pyo.Constraint(model.T, rule=lambda m, t: m.P_charge[t] + m.R_down[t] <= P_max)
    model.reg_up_energy = pyo.Constraint(model.T, rule=lambda m, t: m.E_bat[t] >= (m.R_up[t] * delta_t) / eta_discharge)
    model.reg_down_energy = pyo.Constraint(model.T, rule=lambda m, t: (E_max - m.E_bat[t]) >= (m.R_down[t] * delta_t) * eta_charge)

    # Final SOC constraint (force final SOC to match initial SOC)
    model.E_bat[T-1].fix(E_initial)

    # ==============================
    # Objective Function
    # ==============================
    def obj_rule(m):
        return sum(
            m.P_rt[t] * price_rt[t] * delta_t
            + m.E_da[t] * (price_da[t] - price_rt[t])
            + m.R_up[t]   * price_reg_up[t]
            + m.R_down[t] * price_reg_down[t]
            + m.R_rrs[t]  * price_rrs[t]
            + m.R_ecrs[t] * price_ecrs[t]
            for t in m.T
        )
    model.obj = pyo.Objective(rule=obj_rule, sense=pyo.maximize)

    # ==============================
    # Solve the optimization problem
    # ==============================
    solver = SolverFactory('glpk', executable='/usr/local/bin/glpsol')
    results = solver.solve(model, tee=False)

    # Check solver status
    status = results.solver.status
    term_cond = results.solver.termination_condition

    if (status != pyo.SolverStatus.ok) or (term_cond != pyo.TerminationCondition.optimal):
        return None, None  # No feasible solution

    # Extract results
    total_profit = pyo.value(model.obj)
    final_soc = pyo.value(model.E_bat[T-1])

    return total_profit, final_soc
```

```python
price_rt = DAM_one.SP_Price_Houston.values      # Real-time prices ($/MWh)
price_da = DAM_one.DA_Price_Houston.values      # Day-ahead prices ($/MWh)
price_reg_up = ancilary_service_one.REGUP.values
price_reg_down = ancilary_service_one.REGDN.values
price_rrs = ancilary_service_one.RRS.values
price_ecrs = ancilary_service_one.ECRS.values

profit, final_soc = optimize_battery_schedule(10,price_rt, price_da, price_reg_up, price_reg_down, price_rrs, price_ecrs)
print(f"Total 24-hour profit: ${profit:,.2f}")
print(f"Final State of Charge: {final_soc:.2f} MWh")
```
